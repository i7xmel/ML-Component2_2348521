```python
HappinessScore_bin_edges = [4,6,7, np.inf]
HappinessScore_bin_labels = ['Low-Happiness','Medium-Happiness','High_Happiness']
```

```python
df1 = df.copy()
df1['Happiness_Category'] = pd.cut(df['Happiness Score'], bins=HappinessScore_bin_edges, labels=HappinessScore_bin_labels)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Label_Encoder = LabelEncoder()
df_copy = df.copy()
df_copy['Region'] = Label_Encoder.fit_transform(df_copy['Region'])
```

```python
df_copy
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
new_col=['Economy (GDP per Capita)','Health (Life Expectancy)']
df[new_col] = scaler.fit_transform(df[new_col])
```

```python
df[new_col]
```

```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df_copy['Region'], df_copy['Happiness Rank'])

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Value:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
print(expected)
```

```python
# Covariance matrix of various Categorical data

Economy_Health = df_copy[['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)',]]
cov_matrix1 = np.cov(Economy_Health, rowvar=False)

cov_matrix1
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(cov_matrix1, annot=True, cmap='coolwarm', center=0)
plt.title('Covariance Matrix of GDP, Family and Health ')
plt.show()
```
